```python
from transformers import pipeline, AutoModelForCausalLM

MODEL = 'beomi/KoAlpaca-Polyglot-5.8B'

model = AutoModelForCausalLM.from_pretrained(
    MODEL,
    torch_dtype=torch.float16,
    low_cpu_mem_usage=True,
).to(device=f"cuda", non_blocking=True)
model.eval()

pipe = pipeline(
    'text-generation', 
    model=model,
    tokenizer=MODEL,
    device=0
)
```

```python
import json

# JSON 파일에서 이전 질문과 답변을 로드
try:
    with open('KoAlpaca를_이용한_질문과_답변.json', 'r', encoding='utf-8') as json_file:
        qa_data = json.load(json_file)
except FileNotFoundError:
    qa_data = []

def ask_and_save(questions):
    for question in questions:
        ans = pipe(
            f"### 질문: {question}\n\n### 답변:",
            do_sample=True,
            max_new_tokens=512,
            temperature=0.7,
            top_p=0.9,
            return_full_text=False,
            eos_token_id=2,
        )

        generated_text = ans[0]['generated_text']

        # 생성된 질문과 답변을 JSON 데이터에 추가
        qa_data.append({
            'instruction': question,
            'output': generated_text
        })

        print(f"질문: {question}\n답변: {generated_text}\n")

    # JSON 파일에 데이터 저장
    with open('질문과답변.json', 'w', encoding='utf-8') as json_file:
        json.dump(qa_data, json_file, ensure_ascii=False, indent=4)

# 질문 리스트를 사용하여 질문하고 답변을 저장
ask_and_save(questions_list)
```
