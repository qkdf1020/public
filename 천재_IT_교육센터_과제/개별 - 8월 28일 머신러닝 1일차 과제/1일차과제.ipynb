```python
# 결측치를 칼럼의 평균 값으로 대체
data['Age'] = data['Age'].fillna(data['Age'].mean())

# 칼럼별 결측지 비율 확인
null_check_df = data
null_cnt_df = pd.DataFrame(null_check_df.isnull().sum()).rename(columns = {0:'null_count'}).reset_index()
null_cnt_df['null_ratio'] = round(null_cnt_df['null_count'] / len(null_check_df) * 100, 2)
null_cnt_df

# 결측치 비율이 너무 높은 칼럼 삭제
del data['Cabin']

# 칼럼별 결측지 비율 확인
null_check_df = data
null_cnt_df = pd.DataFrame(null_check_df.isnull().sum()).rename(columns = {0:'null_count'}).reset_index()
null_cnt_df['null_ratio'] = round(null_cnt_df['null_count'] / len(null_check_df) * 100, 2)
null_cnt_df

# 결측치 비율이 아주 작은 칼럼 결측치가 있는 행 제거
data = data.dropna(axis = 0, subset = ['Embarked']) 

# 칼럼별 결측지 비율 확인
null_check_df = data
null_cnt_df = pd.DataFrame(null_check_df.isnull().sum()).rename(columns = {0:'null_count'}).reset_index()
null_cnt_df['null_ratio'] = round(null_cnt_df['null_count'] / len(null_check_df) * 100, 2)
null_cnt_df
```
