```python
# 모듈 불러오기
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Data 지정하기
num = 50

random_var_1 = np.random.randint(0, 100, num)
random_var_2 = np.random.randint(0, 100, num)
random_var_3 = np.random.randint(0, 100, num)

index = pd.date_range('2022-01-01', periods = num)

df = pd.DataFrame([random_var_1, random_var_2, random_var_3]).T
df.index = index
df.columns = ['random_var_1', 'random_var_2', 'random_var_3']

# 차트 정의 하기
trace_1 = go.Scatter(x = df.index,
                     y = df['random_var_1'],
                     mode = 'lines',
                     name = df.columns[0])

trace_2 = go.Scatter(x = df.index,
                     y = df['random_var_2'],
                     mode = 'lines',
                     name = df.columns[1])

trace_3 = go.Scatter(x = df.index,
                     y = df['random_var_3'],
                     mode = 'lines',
                     name = df.columns[2])

_data = [trace_1, trace_2, trace_3]

# 레이아웃
_layout = go.Layout(title = 'Random Yarlable')

# 객체 만들기
fig = go.Figure(data = _data, layout = _layout)

# 도화지 실행
fig.show()
```

## Input & Output

```python
# 모듈 불러오기
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output

# app 객체생성
app = dash.Dash()

# app 레이아웃 설정
app.layout = html.Div([
    html.H2(id = 'Input_H2',
            children = ['Input']),
    dcc.Input(id = 'input',
              value = '글자를 입력하세요.'),
    html.Br(),
    html.Br(),
    html.H2(id = 'Output_H2',
            children = ['output']),
    html.Div(id = 'div_output',
              children = ['아직 입력받지 못했습니다.']),
])

# 변수
@app.callback(Output(component_id = 'div_output',
                     component_property = 'children'),
              [Input(component_id = 'input',
                     component_property = 'value')])
def update_div(value) :
        return f"입력받은 내용입니다 : {value}"

# app 실행
if __name__ == '__main__' :
    app.run()
```

```python
# 모듈 불러오기
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
import numpy as np
import plotly.graph_objects as go

# app 객체생성
app = dash.Dash()

# app 레이아웃 설정
app.layout = html.Div([
    html.H2(id = 'H2',
            children = ['Input & Output Example 3']),
    dcc.Input(id = 'Input_Number',
                 value = 5),
    html.Br(),
    html.Br(),
    dcc.Graph(id = 'graph'),
])

# 변수
@app.callback(Output('graph', 'figure'),
              [Input('Input_Number', 'value')])
def random_fig(N) :
    # Data 지정하기
    N = int(N)
    random_x = np.random.randint(0, 500, N)
    random_y = np.random.randint(0, 1500, N)
    
    # 차트 정의 하기
    trace = go.Scatter(x = random_x,
                       y = random_y,
                       mode = 'markers',
                       name = 'Random_number')
    
    # 레이아웃
    _layout = go.Layout(title = f"Random Number Graph N :{N}")
    
    # 객체 만들기
    fig = go.Figure(data = trace, layout = _layout)
    
    return fig

# app 실행
if __name__ == '__main__' :
    app.run()
```

## Multiple Input & Output

```python
# 모듈 불러오기
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output

# app 객체생성
app = dash.Dash()

# app 레이아웃 설정
app.layout = html.Div([
    # input_1
    html.Div(id = 'div_1',
            children = [html.H3(id = 'H3_Input_1',
                                children = ['Input_1']),
                        dcc.Input(id = 'Input_1',
                                  value = ['입력값_1'])]),
    # input_2
    html.Div(id = 'div_2',
            children = [html.H3(id = 'H3_Input_2',
                                children = ['Input_2']),
                        dcc.Input(id = 'Input_2',
                                  value = ['입력값_2'])]),
    # output
    html.Div(id = 'div_3',
             children = [html.H3(id = 'H3_output',
                                 children = ['Output']),
                         html.Div(id = 'output',
                                  children = ['입력된 내용이 없습니다.'])])
])

# 변수
@app.callback(Output(component_id = 'output',
                     component_property = 'children'),
              [Input('Input_1', 'value'),
               Input('Input_2', 'value')])
def multiple_caic(value_1, value_2) :
    value_1 = int(value_1)
    value_2 = int(value_2)
    
    return f"{value_1} X {value_2} = {value_1 * value_2} 입니다."

# app 실행
if __name__ == '__main__' :
    app.run()
```

```python
# 모듈 불러오기
import dash
from dash import html
from dash import dcc
from dash.dependencies import Input, Output
import numpy as np
import plotly.graph_objects as go

# app 객체생성
app = dash.Dash()

# app 레이아웃 설정
app.layout = html.Div([
    html.H3(id = 'H3',
            children = ['Input_Number']),
    dcc.Input(id = 'Input_Number',
                 value = 100),
    html.Br(),
    html.Br(),
    # output_1
    html.Div(id = 'output_1',
             children = ['None']),
    # output_2
    dcc.Graph(id = 'output_2',),
])

# 변수
@app.callback([Output('output_1', 'children'),
               Output('output_2', 'figure')],
              [Input('Input_Number', 'value')])
def ouput_fune(value) :
    # Data 지정하기
    value = int(value)
    random_x = np.random.randint(0, 100, value)
    random_y = np.random.randint(0, 1000, value)
    
    x_mean = np.mean(random_x)
    y_mean = np.mean(random_y)
    
    # 차트 정의 하기
    trace = go.Scatter(x = random_x,
                       y = random_y,
                       mode = 'markers',)
    
    # 레이아웃
    _layout = go.Layout(title = f"Random Scatter Chart")
    
    # 객체 만들기
    fig = go.Figure(data = trace, layout = _layout)
    
    mean_result = f"X축 평균 : {x_mean} & Y축 평균 :{y_mean}"
    
    return mean_result, fig

# app 실행
if __name__ == '__main__' :
    app.run()
```
