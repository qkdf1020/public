```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from urllib.request import urlretrieve
import pandas as pd
import os
import time
import random

# 최소 및 최대 대기 시간 설정 (예: 'min_wait초 에서 'max_wait'초 사이의 랜덤 대기)
min_wait_1 = 2 # 최소 대기 시간 (초)
max_wait_1 = 5 # 최대 대기 시간 (초)
min_wait_2 = 3 # 최소 대기 시간 (초)
max_wait_2 = 4 # 최대 대기 시간 (초)

# 랜덤 대기 시간 생성
random_wait_1 = random.uniform(min_wait_1, max_wait_1)
random_wait_2 = random.uniform(min_wait_2, max_wait_2)

ind = 1 # 복사할 번호 컨트롤
ind_ = 40 # 현재 페이지에 결번이 있으면 결번 건너 뛰고 지정 번호 까지 시도하고 다음 페이지로 넘어감
no = 1 # 페이지 번호 컨트롤
page_n = 1 # 현재 페이지 번호
page_l = 20 # 페이지 리미트 설정
list_1 = [] # 결과물이 저장되는 리스트

driver = webdriver.Chrome()
driver.get("https://auto.danawa.com/usedcar/?Work=list&Tab=list")

time.sleep(5) # 브라우저가 열릴 떄까지 대기

# 이미지 저장할 폴더 생성
save_folder = os.path.join(os.getcwd(), "danawa")
os.makedirs(save_folder, exist_ok = True)

print(f"-----<<<<< {page_n}에서 {page_l}페이지 까지 다운로드를 시작합니다. >>>>>-----")

while True:
    try:
        img_element = driver.find_element(by = 'xpath', value = f'//*[@id="usedcarList"]/ul/li[{ind}]/div[1]/a/img')
        img_src = img_element.get_attribute("src")
        img_alt = img_element.get_attribute("alt")
        list_1 = [img_src, img_alt]
        
        try:
            # file_name = f"{page_n}-{ind}-{list_1[1]}.jpg"
            file_name = f"{page_n}-{ind}.jpg"
            file_path = os.path.join(save_folder, file_name)
            urlretrieve(list_1[0], file_path)
            print(f"다운로드 완료: {file_name}")
            
        except Exception as e:
            print(f"다운로드 중 오류 발생: {e}")
        
        time.sleep(random_wait_1) # import time
        ind += 1

    except:
        if ind < ind_: 
            # 스크롤 내리기 (페이지 다운 키를 사용)
            driver.find_element(by = By.TAG_NAME, value = "body").send_keys(Keys.PAGE_DOWN)
            ind += 1
            continue
        
        else:
            # 다음 페이지로 넘어가기
            try:
                print(f"-----<<<<< {page_n} 페이지의 다운로드를 완료하였습니다. >>>>>-----")
                if page_n >= page_l:
                    break
                else :
                    no += 1
                    page_n += 1
                    ind = 1
                    driver.find_element(by = 'xpath', value = f'//*[@id="page_area"]/a[{no}]').click()
                    # 페이지 10의 자리가 바뀌서 넘어가면 다시 리셋
                    if no == 11:
                        no = 2

                time.sleep(random_wait_1 + random_wait_2) # import time

            except:
                break
        
```
